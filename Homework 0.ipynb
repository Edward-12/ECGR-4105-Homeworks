```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Problem 1

```python
df = pd.read_csv("./D3.csv")
```

```python
theta = np.zeros((2,1))
print(theta)
```

```python
X1 = df.values[:,0]
X2 = df.values[:,1]
X3 = df.values[:,2]
y = df.values[:,3]

m = len(X1)
X1 = X1.reshape(m,1)
X2 = X2.reshape(m,1)
X3 = X3.reshape(m,1)
y = y.reshape(m,1)

X0 = np.ones((m,1))
X1 = np.hstack((X0, X1))
X2 = np.hstack((X0, X2))
X3 = np.hstack((X0, X3))
```

```python
def compute_loss(X, y, theta):

    h = X.dot(theta)
    error = np.subtract(h, y)
    sqrError = np.square(error)
    J = 1 / (2*m) * np.sum(sqrError)

    return J
```

```python
cost1 = compute_loss(X1, y, theta)
print(cost1)
```

```python
cost2 = compute_loss(X2, y, theta)
print(cost2)
```

```python
cost3 = compute_loss(X3, y, theta)
print(cost3)
```

```python
def gradient_descent(X, y, theta, alpha, iterations):
    
    
    loss_history = np.zeros(iterations)
    
    for i in range(iterations):
        
        h = X.dot(theta)
        error = np.subtract(h, y)
        sum_delta = (alpha / m) * X.transpose().dot(error);
        theta = theta - sum_delta;
        loss_history[i] = compute_loss(X, y, theta)
    return theta, loss_history
```

# Alpha = 0.1

```python
#theta = [0., 0.]
theta = np.zeros((2,1))
print(theta)
iterations = 2000;
alpha = .1;
```

```python
theta, loss_history = gradient_descent(X1, y, theta, alpha, iterations)
total_loss = loss_history.sum()
print(total_loss)
print('Final value of theta = ', theta)
print('Cost_history =', loss_history)
plt.scatter(X1[:,1], y, color='red', marker = '+', label = 'Training data')
plt.plot(X1[:,1], X1.dot(theta), color='green', label='Line Regression')
plt.grid()
plt.xlabel('X1 explanatory variable')
plt.ylabel('y output')
plt.title('X1, y Linear Regression Fit')
plt.legend()
```

```python
plt.plot(range(1, iterations + 1), loss_history, color='magenta')
plt.grid()
plt.xlabel('Number of Iterations')
plt.ylabel('Cost/Loss')
plt.title('Cost vs Iterations')
```
